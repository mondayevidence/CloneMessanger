```python
import networkx as nx

G=nx.Graph()
```

```python
print(G.nodes())
print(G.edges())

print(type(G.nodes()))
print(type(G.edges()))
```

```python
G.add_node("a")
G.add_nodes_from(["b","c"])

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
G.add_edge(1,2)
edge=("d","e")
G.add_edge(*edge)
edge=("a",'b')
G.add_edge(*edge)

print("Nodes of graph: ")
print(G.nodes())
print("Edges of graph: ")
print(G.edges())
```

```python
# Adding a list of edges
G.add_edges_from([('a','c'),('c','d'),('a',1),('1','d'),('a',2)])

import matplotlib.pyplot as plt
nx.draw_networkx(G)
plt.savefig('simple_path.png')
plt.show()
```

```python
G=nx.path_graph(4)

print('Nodes of graph: ')
print(G.nodes())
print('Edges of graph: ')
print(G.edges())
nx.draw_networkx(G)
plt.savefig('path_graph1.png')
plt.show()
```

```python
G=nx.path_graph(4)
cities= {0:'Toronto', 1:'London', 2:'Berlin', 3:'New York'}

H=nx.relabel_nodes(G,cities)

print('Nodes of graph: ')
print(H.nodes())
print('Edges of graph: ')
print(H.edges())
nx.draw_networkx(H)
plt.savefig('path_graph_cities.png')
plt.show()
```
